# Colab egress validation

relay egress is a *hypothesis* until this runs):

1. **HTTPS** to the broker + analyzer — the `--estimate` cell below;
2. **gRPC (TLS, port 9001)** to the relay — the real submit: E2E payload
   delivery and live streaming go through the relay, so a completed task
   with streamed output proves the egress.

```python
size = 512
```

```python
%%krauncher --estimate
import numpy as np
a = np.random.rand(size, size)
mean = float((a @ a).mean())
```

```python
%%krauncher --timeout 300
import numpy as np
a = np.random.rand(size, size)
mean = float((a @ a).mean())
print(f"streamed from the worker: mean={mean:.4f}")
```
